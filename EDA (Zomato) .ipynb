```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observation

1. When ratings between 4.5 to 4.9---> Excelent
2. Ratings between 4.0 to 3.4---> Very good
3. Ratings between 3.5 to 3.9 ----> Good
4. Ratings between 3.0 to 3.4 -----> Average
5. Ratings between 2.5 to 2.9 -----> Average
5. Ratings between 2.0 to 2.4 ------> Very poor
